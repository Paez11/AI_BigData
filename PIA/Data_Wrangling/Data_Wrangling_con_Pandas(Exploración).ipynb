```python
data = pd.read_csv('MailCustomer.csv', index_col = 'CustomerKey', sep = ';' , encoding = "ISO-8859 -1")
print ( data.head() )
print ( " - - - - - - - - - - - - - - - - - - - - " )
print ( data.tail() )
```

La primera vez que abrimos el archivo utilizando la librerÃ­a de pandas, podemos comprobar que se trata de una lista de clientes
los cuales tenemos distintos datos de cada uno de ellos. La mayorÃ­a de estos datos son comunes. Como pueden ser el nombre,
apellidos, direcciÃ³n, telÃ©fono, etc.

```python
print( data.iloc[0]) # Muestra la primera fila
print( data.iloc[: ,0]) # Muestra la primera columna
print( data.iloc[0:5]) # Primeras cinco filas
print( data.iloc[: , 0:5]) # Primeras cinco columnas
print( data.iloc[[0 ,2 ,1]]) # Primera , tercera y segunda filas
```

```python
print( data.columns) # Lista con los nombres de las columnas
print( data.info()) # Informacion util de los datos , como el tipo de datos
```

```python
print(data[['FirstName' , 'LastName']]) # Acceso al nombre y apellido
print(data.loc[: , ['FirstName', 'LastName']]) # Acceso al nombre y apellido
#Si queremos saber cuantos valores distintos tiene cada una de nuestras columnas, podemos hacer lo siguiente:
for i in data . columns :
  print(i ," -->" , len ( data[i].unique().tolist()))
```

ComprobaciÃ³n de valores nulos

```python
print(data.isnull().sum().sort_values(ascending = False))
```

```python
processedData = data.fillna(" None ")
print(processedData.isnull().sum().sort_values(ascending = False)) # Comprobamos que ya no hay valores nulos
```
