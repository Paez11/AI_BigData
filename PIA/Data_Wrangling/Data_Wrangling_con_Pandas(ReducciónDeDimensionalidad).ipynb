```python
data = pd.read_csv('MailCustomer.csv', index_col = 'CustomerKey', sep = ';' , encoding = "ISO-8859 -1")
print ( data.head() )
print ( " - - - - - - - - - - - - - - - - - - - - " )
print ( data.tail() )
```

```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

if 'BirthDate' in data.columns:
  data['BirthDate'] = pd.to_datetime(data['BirthDate']).dt.strftime('%m/%d/%Y')
print(data.BirthDate)

for index, row in data.iterrows():
  birthdate = pd.to_datetime(row['BirthDate'])
  if birthdate.year > 2024:
      birthdate_str = row['BirthDate'] #pasamos la columna
      year = int(birthdate_str[-4:]) #contamos solo el año
      if year > 2024:
          new_year = year - 100
          new_birthdate_str = birthdate_str[:-4] + str(new_year)
          data.loc[index, 'BirthDate'] = new_birthdate_str
print(data.BirthDate)

for i, row in data.iterrows():
  birthdate = pd.to_datetime(row['BirthDate'])
  birthdate_str = row['BirthDate']
  birthyear = int(birthdate_str[-4:])
  updateAge = 2024 - birthyear
  data.loc[i, 'Age'] = updateAge
print(data.Age)

#imputer = SimpleImputer(missing_values = np.nan , strategy = 'constant', fill_value = 'None')
#imputer.fit_transform(data)
data = data.fillna(' None ')

data = data.drop_duplicates()
print(f"Número de registros duplicados eliminados: {data.duplicated().sum()}")

for i, row in data.iterrows():
  birthdate_str = str(row['BirthDate'])
  birthmonth = (birthdate_str[:2])
  birthday = (birthdate_str[3:5])
  birthyear = (birthdate_str[-4:])
  data.loc[i, 'BirthDateDay'] = birthday
  data.loc[i, 'BirthDateMonth'] = birthmonth
  data.loc[i, 'BirthDateYear'] = birthyear

data.drop(['BirthDate'] , axis = 'columns')

Q1 = np.quantile(data['Age'], 0.25)
Q3 = np.quantile(data['Age'], 0.75)
IQR = Q3 - Q1
k = 1.5
xL = Q1 - k * IQR
xU = Q3 + k * IQR
print(f"Bandas: [{xL},{xU}]")
for i in data['Age']:
    if i < xL  or i > xU:
        print(f" El dato[{i}] es un outlier")
        data = data.drop(data[data['Age'] == i].index)

bins = [40, 61, 81, 100] #número de veces que vamos a partir los datos
labels = ['40-60', '61-80', '81+']

data['AgeCategory'] = pd.cut(data['Age'], bins=bins, labels=labels, include_lowest=True)

#print(data[['Age', 'AgeCategory']])
```

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold = 0.25) # Umbral de varianza para seleccionar caracteristicas
print(data.info())
newData = OrdinalEncoder().fit_transform(data[['Title','FirstName','MiddleName','LastName']])
sel = selector.fit(newData)
print(sel.get_support())
```

análisis de correlaciones, de manera que dos características (columnas) que estén muy correladas no tienen que utilizarse conjuntamente

```python
from sklearn.preprocessing import Normalizer

newData = OrdinalEncoder().fit_transform(data[['EnglishOccupation','SpanishOccupation','FrenchOccupation']])
normalizedData = Normalizer().fit_transform(newData)
print(pd.DataFrame(normalizedData).corr(method = 'pearson'))
```

Como podemos ver en el análisis de la correlación anterior, los valores no son exactamente 1 o -1. Pues en el dataset hay un sesgo con las traducciones de la ocupación y la educación por lo que si miramos los datos, aunque estos sean iguales la correlación nos dará un valor distinto por la traducción. Podemos arreglar esto asignando una variable numérica a los valores que son iguales en traducción.

```python
newEnglishOccupation = []
newSpanishOccupation = []
newFrenchOccupation = []

for i in data['EnglishOccupation']:
  if i == 'Clerical':
    newEnglishOccupation.append(1)
  elif i == 'Management':
    newEnglishOccupation.append(2)
  elif i == 'Manual':
    newEnglishOccupation.append(3)
  elif i == 'Professional':
    newEnglishOccupation.append(4)
  elif i == 'Skilled Manual':
    newEnglishOccupation.append(5)

for i in data['SpanishOccupation']:
  if i == 'Administrativo':
    newSpanishOccupation.append(1)
  elif i == 'Obrero':
    newSpanishOccupation.append(3)
  elif i == 'Profesional': #Cadre
    newSpanishOccupation.append(4)
  elif i == 'Obrero especializado': #Technicien
    newSpanishOccupation.append(5)
  else:
    newSpanishOccupation.append(2)

for i in data['FrenchOccupation']:
  if i == 'Direction':
    newFrenchOccupation.append(2)
  elif i == 'Ouvrier':
    newFrenchOccupation.append(3)
  elif i == 'Cadre':
    newFrenchOccupation.append(4)
  elif i == 'Technicien':
    newFrenchOccupation.append(5)
  else:
    newFrenchOccupation.append(1)

data['NewEnglishOccupation'] = newEnglishOccupation
data['NewSpanishOccupation'] = newSpanishOccupation
data['NewFrenchOccupation'] = newFrenchOccupation
data
```

```python
newData1 = OrdinalEncoder().fit_transform(data[['NewEnglishOccupation','NewSpanishOccupation','NewFrenchOccupation']])
normalizedData = Normalizer().fit_transform(newData1)
print(pd.DataFrame(normalizedData).corr(method = 'pearson'))
```

De esta manera confirmamos que los valores son iguales

Además, es posible utilizar la información mutua para ver si dos variables comparten información o son totalmente independientes.

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import OrdinalEncoder
targetData = data ['EnglishOccupation']
newData = OrdinalEncoder().fit_transform(data[['EnglishOccupation','SpanishOccupation','FrenchOccupation']])
importances = mutual_info_classif(newData, targetData)
print(importances)

targetData1 = data ['NewEnglishOccupation']
newData1 = OrdinalEncoder().fit_transform(data[['NewEnglishOccupation','NewSpanishOccupation','NewFrenchOccupation']])
importances1 = mutual_info_classif(newData1, targetData1)
print(importances1)
```

Si queremos incluso reducirlo aun más. Podemos incluso eliminar datos que aunque pueden aportar algo tampoco son esenciales, pues podemos sacar el conocimiento a partir de otros datos.

```python
data = data.drop(['CustomerAlternateKey', 'AddressLine1', 'EmailAddress', 'Phone'], axis = 'columns')
data.head()
```

```python

data['DateFirstPurchase'] = data['DateFirstPurchase'].str.extract(r'(\d{2,4})$')  # Extraer el año de 4 dígitos
data['DateFirstPurchase'] = data['DateFirstPurchase'].astype(float).dropna().astype(int)  # Convertir a entero y eliminar nulos

def formattedYear(year):
    if year < 100:  # Si el año tiene 2 dígitos
        return 2000 + year if year <= 20 else 1900 + year
    return year

data['DateFirstPurchase'] = data['DateFirstPurchase'].apply(formattedYear)

data.head()
```
