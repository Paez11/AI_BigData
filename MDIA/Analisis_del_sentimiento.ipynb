```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
# Cargar el modelo y el tokenizer
model_name = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
def analyze_sentiment(text):
    # Preprocesar el texto (tokenización)
    encoded_input = tokenizer(text, return_tensors='pt')

    # Hacer la predicción
    output = model(**encoded_input)
    scores = output.logits.detach().numpy()[0]

    # Aplicar softmax para convertir a probabilidades
    probabilities = softmax(scores)

    # Etiquetas de sentimiento
    labels = ["Negative", "Neutral", "Positive"]

    # Crear un diccionario con los resultados
    results = {labels[i]: float(probabilities[i]) for i in range(len(labels))}

    return results
```

```python
# Ejemplo de uso
tweet = "¡Me encanta cómo está avanzando la tecnología!"
result = analyze_sentiment(tweet)
print(result)
```
